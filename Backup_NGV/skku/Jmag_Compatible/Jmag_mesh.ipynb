```python

void readGroupInCondition(JcfDataPtr data)
{
	int id = 1;
	const char* condName = "Coil_in_FEMCoil";
	long check;	/* SimpleFHI API�̐����E���s�̏�� */
	int numGrpus;
	int i, j;
	int elems;
	int type;
	long first, second;

	/*
		---- FEM�R�C���̃O���[�v�̓ǂݍ��� ----
	 */
	numGrpus = countConditionGroups(data, condName, id);	// �����Ɋ܂܂�Ă���O���[�v��
	/* �\�� */
	printf("groups in condition: %d\n", numGrpus);
	for(i=0; i<numGrpus; ++i)
	{
		type = readConditionGroupType(data, condName, id, i);		// �O���[�v�̃^�C�v
		elems = countConditionGroupElments(data, condName, id, i);	// �O���[�v�Ɋ܂܂�Ă���v�f��
		/* �\�� */
		printf("group type: %d\n", type);
		printf("group elements: %d\n", elems);
		for(j=0; j<elems; ++j)
		{
			check = readConditionGroup(data, condName, id, i, j, &first, &second);	// �O���[�v�̗v�f
			/* �\�� */
			printf("group element: (%d, %d)\n", first, second);
		}
		printf("\n");
	}
}

/*
	�R���g���[���p�����[�^�̏�������
 */
```
