```python


void writeControlParameter(JcfDataPtr data)
{
	long check;	/* SimpleFHI APIï¿½Ìï¿½ï¿½ï¿½ï¿½Eï¿½ï¿½ï¿½sï¿½Ìï¿½ï¿½ */

	check = writeControlParameterString(data, "common", "title", "Modified Title");	/* ï¿½^ï¿½Cï¿½gï¿½ï¿½ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
    check = writeControlParameterInteger(data, "common", "max_plot_size", 50);	/* ï¿½vï¿½ï¿½ï¿½bï¿½gï¿½tï¿½@ï¿½Cï¿½ï¿½ï¿½Ì•ï¿½ï¿½ï¿½ï¿½Tï¿½Cï¿½Yï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
}


/*
	ï¿½Ş—ï¿½ï¿½pï¿½ï¿½ï¿½ï¿½ï¿½[ï¿½^ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½
 */
void writeMaterialParameter(JcfDataPtr data)
{
	long check;	/* SimpleFHI APIï¿½Ìï¿½ï¿½ï¿½ï¿½Eï¿½ï¿½ï¿½sï¿½Ìï¿½ï¿½ */

	/*
		---- ï¿½Ş—ï¿½(ï¿½|ï¿½ï¿½)ï¿½Ì“Ç‚İï¿½ï¿½ï¿½ ----
	 */
	int id = 1;
	check = writeMaterialParameterInteger(data, id, "coordinate_type", 1);	/* ï¿½ï¿½ï¿½Wï¿½nï¿½^ï¿½Cï¿½vï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */

	/*
		---- ï¿½Ş—ï¿½(ï¿½ï¿½ï¿½ï¿½)ï¿½Ì“Ç‚İï¿½ï¿½ï¿½ ----
	 */
	id = 2;
    check = writeMaterialParameterDouble(data, id, "temperature", 100.0);	/* ï¿½ï¿½ï¿½xï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
	check = writeMaterialParameterInteger(data, id, "pattern_type", 3);	/* ï¿½ï¿½ï¿½ï¿½ï¿½pï¿½^ï¿½[ï¿½ï¿½ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
}


/*
	ï¿½ï¿½ï¿½ï¿½ï¿½pï¿½ï¿½ï¿½ï¿½ï¿½[ï¿½^ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½
 */
void writeConditionParameter(JcfDataPtr data)
{
	int id = 1;	/* ï¿½ï¿½ï¿½ï¿½ */
	long check;	/* SimpleFHI APIï¿½Ìï¿½ï¿½ï¿½ï¿½Eï¿½ï¿½ï¿½sï¿½Ìï¿½ï¿½ */

	/*
		---- FEMï¿½Rï¿½Cï¿½ï¿½ï¿½Ì“Ç‚İï¿½ï¿½ï¿½ ----
	 */
    check = writeConditionParameterDouble(data, "FEMCoil", id, "coil_zlength", 0.05);	/* ï¿½Rï¿½Cï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
    check = writeConditionParameterString(data, "FEMCoil", id, "fem_coil_name_gui", "Modified Name");	/* FEMï¿½Rï¿½Cï¿½ï¿½ï¿½ï¿½ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
}


int main(int argc, char** argv)
{
	JcfDataPtr data;
	long check;

	/*
		arg[1] : ï¿½ï¿½ï¿½Íƒtï¿½@ï¿½Cï¿½ï¿½ï¿½ï¿½
		arg[2] : ï¿½oï¿½Íƒtï¿½@ï¿½Cï¿½ï¿½ï¿½ï¿½
	 */

	if(argc!=3)
	{
		fprintf(stderr, "parameter error\n");
		return EXIT_FAILURE;
	}

    /* jcfï¿½tï¿½@ï¿½Cï¿½ï¿½ï¿½ï¿½ï¿½Jï¿½ï¿½ */
    check = loadJcfFile(argv[1], &data);
	if(check==0)
	{
		fprintf(stderr, "Can not open file: %s\n", argv[1]);
		return EXIT_FAILURE;
	}
       
	/*
		---- ï¿½Ç‚İï¿½ï¿½İ—ï¿½ ----
	 */
    /* ï¿½Rï¿½ï¿½ï¿½gï¿½ï¿½ï¿½[ï¿½ï¿½ï¿½pï¿½ï¿½ï¿½ï¿½ï¿½[ï¿½^ï¿½Ì“Ç‚İï¿½ï¿½ï¿½ */
    readControlParameter(data);

	/* ï¿½Ş—ï¿½ï¿½pï¿½ï¿½ï¿½ï¿½ï¿½[ï¿½^ï¿½Ì“Ç‚İï¿½ï¿½ï¿½ */
	readMaterial(data);

	/* ï¿½ï¿½ï¿½ï¿½ï¿½pï¿½ï¿½ï¿½ï¿½ï¿½[ï¿½^ï¿½Ì“Ç‚İï¿½ï¿½ï¿½ */
	readCondition(data);

	/* ï¿½ï¿½ï¿½ï¿½ï¿½ÉŠÜ‚Ü‚ï¿½ï¿½Oï¿½ï¿½ï¿½[ï¿½vï¿½Ì“Ç‚İï¿½ï¿½ï¿½*/
	readGroupInCondition(data);

	/*
		---- ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ ----
	 */
	/* ï¿½Rï¿½ï¿½ï¿½gï¿½ï¿½ï¿½[ï¿½ï¿½ï¿½pï¿½ï¿½ï¿½ï¿½ï¿½[ï¿½^ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
	writeControlParameter(data);

	/* ï¿½Ş—ï¿½ï¿½pï¿½ï¿½ï¿½ï¿½ï¿½[ï¿½^ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
	writeMaterialParameter(data);
	
	/* ï¿½ï¿½ï¿½ï¿½ï¿½pï¿½ï¿½ï¿½ï¿½ï¿½[ï¿½^ï¿½Ìï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ï¿½ */
	writeConditionParameter(data);

	/* jcfï¿½tï¿½@ï¿½Cï¿½ï¿½ï¿½ğ–¼‘Oï¿½ï¿½tï¿½ï¿½ï¿½Ä•Û‘ï¿½ */
	saveAsJcfFile(data, argv[2]);

	/* jcfï¿½tï¿½@ï¿½Cï¿½ï¿½ï¿½ï¿½Â‚ï¿½ï¿½ï¿½ */
	closeJcfFile(data);

	return EXIT_SUCCESS;
}
```
